```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv(r'D:\MLProject\cbb18.csv')
data.head()
```

```python
data.describe(include='all')
```

```python
data_num = data[data.select_dtypes(include=[np.number]).columns.tolist()]
```

```python
data_num.head()
```

```python
data_dummies = pd.get_dummies(data[data.select_dtypes(include=['object']).columns.tolist()], drop_first=True)
```

```python
data_dummies.head()
```

```python
data_combined = pd.concat([data_num, data_dummies], axis=1)
```

```python
data_combined.head()
```

```python
from sklearn.model_selection import train_test_split

X = data_combined.drop(['POSTSEASON_Champions'],axis=1)
Y = data_combined['POSTSEASON_Champions']

X.fillna(X.mean(), inplace=True)

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3,random_state=20)
X_train.columns
```

```python
from sklearn.ensemble import RandomForestClassifier


training = RandomForestClassifier(n_estimators=100, bootstrap= False, criterion= 'entropy', 
                            max_depth= None, max_features= None, min_samples_split= 6)


training.fit(X_train, y_train)


y_pred = training.predict(X_test)
```

```python
training
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

print("Precision:",metrics.precision_score(y_test, y_pred))

print("Recall:",metrics.recall_score(y_test, y_pred))
```
